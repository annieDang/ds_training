### Step 2: Data cleaning

* <strong>cut off unrelavant data that out

```python
# Since we're only focusing on millennial generation age group (23 to 38), 
# all other demographical groups are removed
census_income_data = census_income_data.drop(census_income_data[census_income_data.age < 23].index)
census_income_data = census_income_data.drop(census_income_data[census_income_data.age > 38].index)
```

```python
# test the removal
print census_income_data['age'].aggregate(["min", "max"])
print census_income_data.shape
```

```python
# delete unrelevant columns
del census_income_data['na']
```

* <strong>Handle missing value, whitespace

```python
# remove start white space
categorical_data = census_income_data.select_dtypes(include='O')
for column in categorical_data:
    census_income_data[column] = census_income_data[column].str.strip()
```

```python
print census_income_data.isna().any().sum()
```

```python
# remove unknow major industry code
unknow_major_industry_code = census_income_data.major_industry_code=='Not in universe or children'
census_income_data = census_income_data.drop(census_income_data[unknow_major_industry_code].index)
print census_income_data.shape
```

* <Strong> brief look after all cleaning

```python
# A brief look at all categorical data again
categorical_data = census_income_data.select_dtypes(include='O')
for column in categorical_data:
    
    counts = categorical_data.groupby(column).size() 
    print counts
    print(census_income_data[column].unique())
    print '*' * 50
```

```python
# A brief look at all numerical data
numerical_data = census_income_data.select_dtypes(include=['int64','float'])
aggregate = numerical_data.aggregate(["min", "max"])
for column in numerical_data:
    print(column + ":"  )
    print '_' * 20
    print aggregate[column]
    print '*' * 50
```

## 2.1 Exploring interested columns

* <strong> a quick look at demographic

>  some interested categorical fields : sex, race, education, etc.

```python
groupNames = ['sex', 'race', 'education', 'country_of_birth_self', 'marital_stat', 'major_industry_code', 'own_business_or_self_employed']
for each in groupNames:
    grouped = census_income_data.groupby(each).size().sort_values(ascending=False)
    grouped.plot(kind='bar', figsize=(10,6), color="blue", fontsize=13);
    plt.show()
    
```

> some interested numerical fields : wage per hour, divdends from stocks,  total person income, etc.

```python
groupNames = ['wage_per_hour', 'dividends_from_stocks', 'age', 'weeks_worked_in_year', 'capital_gains', 'capital_losses']

for each in groupNames:
    plt.hist(census_income_data[each], density=True, bins=50)
    plt.xlabel(each)
    plt.show()
```
